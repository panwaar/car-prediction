# Quality 

1. Year has non-year values (object)
2. Year object to integer
3. Price has "Ask for price" (remove this)
4. Convert Price from object to int
5. kms_driven has kms with values
6. kms_driven has nan values 
7. fuel_type  has nan values
8. keep first 3 words of name

# Cleaning

```python
backup = car.copy()
```

```python
car = car[car['year'].str.isnumeric()]
```

```python
car['year'] = car['year'].astype(int)
```

```python
car = car[car['Price'] !="Ask For Price" ]
```

```python
car['Price'] = car['Price'].str.replace(',','').astype(int)
```

```python
car['kms_driven'] = car['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
car = car[car['kms_driven'].str.isnumeric()]
```

```python
car['kms_driven'] = car['kms_driven'].astype(int)
```

```python
car = car[~car['fuel_type'].isna()]
```

```python
car['name']
```

```python
car['name'] = car['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
car.reset_index(drop=True)
```

```python
car.describe()
```

```python
car = car[car['Price']<6e6].reset_index(drop=True)
```

```python
car
```

```python
car.to_csv('Cleaned car.csv') 
```

# Model

```python
X = car.drop(columns='Price')
y = car['Price']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test , y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python

column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_), ['name','company','fuel_type']),remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
# this pipeline will first go to one hot encoder, encode the data then all these transformer cols along with the previous cols will be given to this linear regression model  
pipe = make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred) 
```

```python
score=[]
for i in range(1000):
    X_train, X_test,y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=i)
    lr = LinearRegression()
    pipe = make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred = pipe.predict(X_test)
    # print(r2_score(y_test,y_pred), i)
    score.append(r2_score(y_test, y_pred))
```

```python
np.argmax(score)
```

```python
score[np.argmax(score)]
# score[800]
```

```python
X_train, X_test,y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=np.argmax(score))
lr = LinearRegression()
pipe = make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
r2_score(y_test, y_pred)
```
